0から1の値をとる一様乱数を1,000個、2組発生させて、それぞれのヒストグラムを描いてみましょう。
なお、それぞれのヒストグラムを別のグラフに表示するために、plt.subplotを利用してください。また、ここで一
様乱数とは、ある数から別のある数まで等確率で発生する乱数のことをいい、np.random.uniformを使います。た
とえば、0から1までの数を10個発生させる場合は、np.random.uniform(0.0, 1.0, 10)とします。
また、1,000個だけではなく、100個や10,000個などでも実施してみましょう。何かわかることはありますか。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
RandomArray1 = np.random.rand(1000)
RandomArray2 = np.random.rand(1000)

# RandomArray1 = np.random.uniform(0.0, 1.0, 1000)
# RandomArray2 = np.random.uniform(0.0, 1.0, 1000)
```

```python
plt.subplot(1, 2, 1)
plt.hist(RandomArray1, bins=30)
plt.title('Random Array 1')

plt.subplot(1, 2, 2)
plt.hist(RandomArray2, bins=30)
plt.title('Random Array 2')
```

```python
RandomArrayLen = 10

RandomArray1 = np.random.rand(RandomArrayLen)
RandomArray2 = np.random.rand(RandomArrayLen)

# RandomArray1 = np.random.uniform(0.0, 1.0, RandomArrayLen)
# RandomArray2 = np.random.uniform(0.0, 1.0, RandomArrayLen)

plt.subplot(1, 2, 1)
plt.hist(RandomArray1, bins=30)
plt.title('Random Array 1')

plt.subplot(1, 2, 2)
plt.hist(RandomArray2, bins=30)
plt.title('Random Array 2')
```
