標準正規分布に従う乱数を10個発生させて配列を作成してください。また、その中での最小値、最大値、合計を
求めるプログラムを書いてください。

```python
import numpy as np
```

```python
RandomList = [np.random.normal() for i in range(10)]
```

```python
print('Max:' + str(max(RandomList))
      + '\nMin:' + str(min(RandomList))
      + '\nSum:' + str(np.sum(RandomList)))
```
