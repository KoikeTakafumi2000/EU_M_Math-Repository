回帰：統計学において連続したyの値がある時にy=f(x)と当てはめること．

分類：様々な属性を持つものをその属性ごとに分けること．

教師あり学習：正解(教師)があり，正解率が高くなる，あるいは正解との誤差が少なくなるように学習させること

重回帰分析：回帰分析の中で目的変数に対して複数の説明変数があるもの

ロジスティクス回帰分析：回帰分析の中で正解がbool値で示されるもの．確率を0から1で推定し，誤差を減らす学習をさせる．

正則化：モデルの複雑さを低減する工夫のこと．複雑なモデルを許容すると過学習に陥りやすいため，重みなどが大きくなるほど誤差にプラスするなどして，複雑なモデルを許容しないシステムを作る．

リッジ回帰：損失関数に最小二乗法にプラスして各項の重みの二乗の合計を既定の倍率を掛けて足すことで，重みが大きな項(複雑な回帰式)の出現を抑制する回帰
ラッソ回帰よりもより大きな重みに対する罰則が強い

ラッソ回帰：損失関数に最小二乗法にプラスして各項の重みの絶対値の合計を既定の倍率を掛けて足すことで，重みが大きな項(複雑な回帰式)の出現を抑制する回帰
リッジ回帰よりも小さな重みにも罰則をより多く着ける

決定木：特定の条件で分岐し，分岐によって分類する．同じデータ，決定木を使えば乱数に左右されず同じ結果が出る．なぜその結果が出たのかを説明できるという特徴がある．
分類することでエントロピーが削減されるように分類を設定する．分類によって削減されるエントロピーを情報利得と呼ぶ．

エントロピー：乱雑さ，正解ラベルの偏りが少ないとエントロピーが少なく，均等に混ざっている場合はエントロピーが大きい．

情報利得：分類することで削減できたエントロピーのこと

k-NN法：分類されたデータに新しく分類されていないデータが入ってきた際に，半径kの範囲に最も存在するものをその新しく来たデータの分類にする方法．

SVM：サポートベクターマシン．カテゴリを分類する境界線を引く手法．それぞれのカテゴリに最も近い点との距離が最大になるように境界線を引く．

ノーフリーランチ：うまい話は無いということ．統計分野においては範化性能と特定の問題に対する性能はトレードオフの関係にあり，特定の問題に強いモデルでも，範化性能は低い．そのためすべてを満たす条件のいいモデルは存在しない．範化性能が高いモデルでも，得意分野があり，すべてにおいて範化であるわけではない．
