下記の2つのデータテーブルに対して、内部結合してみましょう。

```python
import numpy as np
import pandas as pd
```

```python
# データ 4 の準備
data4 = {
'id': ['0', '1', '2', '3', '4', '6', '8', '11', '12', '13'],
'city': ['Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo', 'Tokyo', 'Osaka', 'Kyoto',
'Hokkaido', 'Tokyo'],
'birth_year': [1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
'name': ['Hiroshi', 'Akiko', 'Yuki', 'Satoru', 'Steeve', 'Mituru', 'Aoi', 'Tarou', 'Suguru',
'Mitsuo']
}
df4 = pd.DataFrame(data4)
df4
```

```python
# データ 5 の準備
data5 = {
'id': ['0', '1', '3', '6', '8'],
'math' : [20, 30, 50, 70, 90],
'english': [30, 50, 50, 70, 20],
'sex': ['M', 'F', 'F', 'M', 'M'],
'index_num': [0, 1, 2, 3, 4]
}
df5 = pd.DataFrame(data5)
df5
```

```python
pd.merge(df4, df5, on='id', how='inner')
```
