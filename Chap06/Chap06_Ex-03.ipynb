```python
import pandas as pd
import numpy as np
```

```python
hier_df1 = pd.DataFrame(
    np.arange(12).reshape((3,4)),
    index = [['c','d','d'],[1,2,1]],
    columns = [
        ['Kyoto','Nagoya','Hokkaido','Kyoto'],
        ['Yellow','Yellow','Red','Blue']
    ]
    )
hier_df1.index.names = ['key1','key2']
hier_df1.columns.names = ['city','color']
hier_df1
```

```python
hier_df1.groupby('key2').mean()
```
