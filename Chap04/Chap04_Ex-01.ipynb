コインの表を0、裏を1として表現する配列を次のように用意します。
coin_data = np.array([0,1])
この配列を使って、コイン投げの試行を1,000 回実行し、その結果、表（値が0）ならびに裏（値が1）が出るそれぞ
れの確率を求めてください。

```python
# 以下のライブラリを使うので、あらかじめ読み込んでおいてください
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
# 可視化ライブラリ
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
# 小数第 3 位まで表示
%precision 3
# ランダムシードの固定
np.random.seed(0)
```

```python
coin_data = np.array([0,1])
```

```python
coin_count = 1000
```

```python
reverse_count = np.sum(np.random.choice(coin_data, coin_count, replace=True))
```

```python
print(f"表が出た回数: {coin_count-reverse_count} 回")
print(f"裏が出た回数: {reverse_count} 回")
print(f"表が出た割合: {(coin_count-reverse_count)/coin_count:.3f}")
print(f"裏が出た割合: {reverse_count/coin_count:.3f}")
```
