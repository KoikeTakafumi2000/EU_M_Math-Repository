3 章で使った、数学の成績を示すデータである「student-mat.csv」を使って、学校を選んだ理由（reason）を円
グラフ化して、それぞれの割合を出してください。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Chap03/student-mat.csv',sep=';')
data.head()
```

```python
data.groupby('higher')['G3'].mean()
```

```python
data.groupby('higher')['G3'].mean()['yes']
```

```python
plt.bar(1, data.groupby('higher')['G3'].mean()['no'])
plt.bar(2, data.groupby('higher')['G3'].mean()['yes'])
```
