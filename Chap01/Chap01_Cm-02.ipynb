```python
N = int(input("Enter a number: "))

PrimeList = []
for i in range(N):
    if i < 2:
        continue
    is_prime = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        PrimeList.append(i)
        
print(PrimeList)
```
