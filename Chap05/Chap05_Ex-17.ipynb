Sicpyを用いて、以下の関数が0となる解を求めましょう。

```python
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 5*x - 10
```

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)), 'r--')
plt.show()
```

```python
fsolve(f,1)
```
