以下の最適化問題をSicpyを使って解いてみましょう。

min f(x) = x^2 + 1
s.t.x >= -1

```python
from scipy.optimize import minimize
```

```python
# 目的となる関数
def objective(x):
    return x*x + 1
```

```python
# 制約式その 1
def constraint1(x):
    return x + 1


# 初期値
x0 = 1
print(objective(x0))
```

```python

con1 = {'type':'ineq','fun':constraint1}
cons = [con1]
```

```python
sol = minimize(objective,x0,method='SLSQP',constraints=cons)
print(sol)
```
