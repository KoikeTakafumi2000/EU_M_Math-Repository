次のx_array、y_arrayがあるとき、Numpyのwhereを用いて条件制御し、3 番目と4 番目はx_arrayから、1 番目、
2 番目、5 番目はy_arrayから、それぞれ値を取り出したデータを生成してください。

```python
import numpy as np
import numpy.random as random
```

```python
x_array= np.array([1,2,3,4,5])
y_array= np.array([6,7,8,9,10])
```

```python
bool_array = np.array([True, True, False, False, True])
```

```python
np.where(bool_array, x_array, y_array)
```
