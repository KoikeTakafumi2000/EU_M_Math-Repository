以下のデータに対して、線形補間の計算をして、グラフを描いてください。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.sin(x**2/5.0)
plt.plot(x,y,'o')
plt.grid(True)
```

```python
f = interpolate.interp1d(x, y,'linear')
```

```python
plt.plot(x,y,'o')
plt.plot(x,f(x),'-')
plt.grid(True)
```
