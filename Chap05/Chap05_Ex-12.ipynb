さらに、3 次元のスプライン補間も加えてみましょう。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.sin(x**2/5.0)
plt.plot(x,y,'o')
plt.grid(True)
```

```python
f = interpolate.interp1d(x, y,'linear')
f2 = interpolate.interp1d(x, y,'quadratic')
f3 = interpolate.interp1d(x, y,'cubic')
```

```python
x = np.linspace(0, 10, num=11, endpoint=True)
x2 = np.linspace(0, 10, num=100, endpoint=True)
y = np.sin(x**2/5.0)
plt.plot(x,y,'o')
plt.plot(x2,f(x2),'-')
plt.plot(x2,f2(x2),'-.')
plt.plot(x2,f3(x2),'--')
plt.grid(True)
```
