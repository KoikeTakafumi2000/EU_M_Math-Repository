右の行列に対して、コレスキー分解を活用して、Ax=bの方程式を解いてください。

```python
import numpy as np
import scipy as sp
```

```python
A = np.array([[5, 1, 0, 1],
              [1, 9, -5, 7],
              [0, -5, 8, -3],
              [1, 7, -3, 10]])
b = np.array([2, 10, 5, 10])
```

```python
b = np.array([5, 20, 0, 20])
L = sp.linalg.cholesky(A)
t = sp.linalg.solve(L.T.conj(), b)
x = sp.linalg.solve(L, t)
# 解答
print(x)
```

```python
(np.linalg.inv(A))@b
```
