```python
# x として、linspace で、開始が 0、終了が 10、項目が 11 つの等間隔数列を生成
x = np.linspace(0, 10, num=11, endpoint=True)
# y の値を生成
y = np.cos(-x**2/5.0)
plt.plot(x,y,'o')
plt.grid(True)
```

```python
from scipy import interpolate
# 線形補間。interp1d のパラメータとして「linear」を指定する
f = interpolate.interp1d(x, y,'linear')
plt.plot(x,f(x),'-')
plt.grid(True)
```

```python
# スプライン 3 次補間を計算して f2 として追加する。パラメータに「cubic」を指定する
f2 = interpolate.interp1d(x, y,'cubic')
# 曲線を出すために、x の値を細かくする。
xnew = np.linspace(0, 10, num=30, endpoint=True)
# グラフ化。f を直線で描き、f2 を点線で描く
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
# 凡例
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.grid(True)
```

```python
# 2 重積分
def I(n):
    return integrate.dblquad(lambda t, x: np.exp(-x*t)/t**n, 0, np.inf, lambda x: 1, lambda x: np.inf)
print('n=1 の時 :',I(1))
print('n=2 の時 :',I(2))
print('n=3 の時 :',I(3))
print('n=4 の時 :',I(4))
```

```python
# モジュールの読み込み
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# ローレンツ方程式
def lorenz_func(v, t, p, r, b):
    return [-p*v[0]+p*v[1], -v[0]*v[2]+r*v[0]-v[1], v[0]*v[1]-b*v[2]]
```

```python
# パラメータの設定
p = 10
r = 28
b = 8/3
v0 = [0.1, 0.1, 0.1]
t = np.arange(0, 100, 0.01)
# 関数の呼び出し
v = odeint(lorenz_func, v0, t, args=(p, r, b))
# 可視化
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(v[:, 0], v[:, 1], v[:, 2])
# ラベルなど
plt.title('Lorenz')
plt.grid(True)
```

```python
from scipy.optimize import minimize
```

```python
# 目的となる関数
def objective(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    return x1*x4*(x1+x2+x3)+x3
```

```python
# 制約式その 1
def constraint1(x):
    return x[0]*x[1]*x[2]*x[3]-25.0
# 制約式その 2
def constraint2(x):
    sum_sq = 40
    for i in range(4):
        sum_sq = sum_sq - x[i]**2
    return sum_sq
# 初期値
x0 = [1,5,5,1]
print(objective(x0))
```

```python
constraint2(x0)
```

```python
b = (1.0,5.0)
bnds = (b,b,b,b)
con1 = {'type':'ineq','fun':constraint1}
con2 = {'type':'ineq','fun':constraint2}
cons = [con1,con2]
```

```python
sol = minimize(objective,x0,method='SLSQP',bounds=bnds,constraints=cons)
print(sol)
```

```python
print('Y:',sol.fun)
print('X:',sol.x)
```
