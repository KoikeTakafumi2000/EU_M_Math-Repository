同様に、以下の関数が0となる解を求めましょう。

```python
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x*x*x - 2*x*x -11*x + 12
```

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)), 'r--')
plt.ylim(-10, 10)
plt.show()
```

```python
print(fsolve(f,-3))
print(fsolve(f,2))
print(fsolve(f,5))
```
