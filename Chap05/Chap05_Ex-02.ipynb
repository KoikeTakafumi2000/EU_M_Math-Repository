練習問題 5-1で使ったデータsample_namesとdataを使って、dataから、sample_namesのc以外に該当するデー
タを抽出してください。

```python
import numpy as np
import numpy.random as random
```

```python
# データの準備
sample_names = np.array(['a','b','c','d','a'])
random.seed(0)
data = random.randn(5,5)
print(sample_names)
print(data)
```

```python
data[sample_names!='c']
```
